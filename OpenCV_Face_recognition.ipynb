## How you can instruct OpenCV to find a face in an image?

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import cv2

img = cv2.imread("foto.jpg")
```

```python
i = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(i)
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, "gray")
plt.show()
```

#### We using a classifier from (https://docs.opencv.org/4.x/d2/d99/tutorial_js_face_detection.html) to recognize faces

```python
classifier = cv2.CascadeClassifier("haarcascade_frontalface_alt2.xml")
```

```python
faces = classifier.detectMultiScale(gray, minNeighbors=10)
```

```python
print(faces)
```

```python
c = img.copy()
for face in faces:
    x, y, w, h = face
    cv2.rectangle(c, (x, y), (x + w, y + h), (0, 255, 0), 2)
    print(face)

i = cv2.cvtColor(c, cv2.COLOR_BGR2RGB)
plt.imshow(i)
plt.show()
```
