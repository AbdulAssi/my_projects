```python
from PIL import Image

img = Image.open("bild1.jpg")
img = img.resize((round(img.size[0] / 10), round(img.size[1] / 10)), resample=Image.BICUBIC)
img
```

```python
img2 = img.crop((50, 50, 150, 150)).resize((32, 32), resample=Image.BICUBIC)
img2.size
```

```python
import numpy as np

data = np.asarray(img2).astype(np.float32) / 255.
data
```

```python
from keras.models import load_model

filename = "cardetector.keras"

model = load_model(filename)
```

```python
print(data.shape)
```

```python
model.predict(data.reshape(1, 32, 32, 3))
```

```python
if model.predict(data.reshape(1, 32, 32, 3)) > 0.90:
    print("There is a Car")
else:
    ("There is no Car")
```

```python
import matplotlib.pyplot as plt

plt.imshow(data)
plt.show()
```
